```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from fpdf import FPDF
```

```python
# Example data
X = np.array([1, 2, 3, 4, 5]) # Perfect positive linear relationship
Z = np.array([-1, -2, -3, -4, -5])  # Perfect negative linear relationship
V = np.array([3, 5, 2, 8, 7])       # No linear relationship
Y = np.array([2, 4, 6, 8, 10])  # target feature

# Calculate Pearson correlation coefficient
pearson_corr, _ = pearsonr(X, Y)
print(f"Pearson correlation coefficient: {pearson_corr}")
```

```python
# Create a scatter plot using Seaborn
sns.regplot(x=X, y=Y, scatter_kws={"s":50}, line_kws={"linewidth": 1.5, "color": "red"})
plt.title('Perfect positive linear relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.savefig('perfect_positive_linear_relationship.png', bbox_inches='tight')
plt.show()
```

```python
sns.regplot(x=Z, y=Y, scatter_kws={"s":50}, line_kws={"linewidth": 1.5, "color": "red"})
plt.title('Perfect negative linear relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
sns.regplot(x=V, y=Y, scatter_kws={"s":50}, line_kws={"linewidth": 1.5, "color": "red"})
plt.title('No linear relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
